**Завдання: Система відстеження витрати пального для техніки**

У вас є таблиця з даними про витрату пального для різної техніки. Таблиця містить наступні поля:

1. `Техніка` - назва або ідентифікатор техніки.
2. `Дата` - дата транзакції.
3. `Кількість літрів` - кількість пального, витраченого в літрах.
4. `Вартість` - вартість заправки в вашій валюті.
5. `Пробіг` - пробіг техніки на момент заправки.

**Частина 1: Розробка структури БД**

Створіть структуру БД для зберігання цих даних у системі PostgreSQL. Створіть дві таблиці: одна для запису даних про витрату пального, а інша для даних про техніку. Кожен запис про витрату пального повинен бути пов'язаний з певним об'єктом техніки.

**Частина 2: Отримання даних**

Напишіть Python back-end, який забезпечить зчитування та запис даних у вашу базу даних PostgreSQL. Розробіть API, яке дозволить:

- Додавати нові записи про витрату пального у базу даних, пов'язуючи їх з конкретним об'єктом техніки.
- Отримувати список всіх записів про витрату пального для певної техніки.
- Отримувати статистику, таку як середній розхід пального на кілометр для конкретної техніки.

**Частина 3: Аналіз даних за допомогою ML**

Використовуючи дані з вашої бази даних, розробіть модель машинного навчання, яка може передбачити середній розхід пального на кілометр для конкретної техніки на основі історичних даних. Виведіть результати прогнозу на вашому веб-дашборді.

**Частина 4: Виведення на дашборд**

Створіть веб-дашборд за допомогою HTML, CSS і JavaScript, який дозволить користувачам візуалізувати дані про витрату пального для різної техніки за допомогою інтерактивних графіків. Ваш дашборд має містити наступні елементи:

- Вибір техніки для відображення даних.
- Графік, що показує залежність середнього розходу від пробігу для вибраної техніки.
- Графік, що відображає зміну витрати на пальне від дати для вибраної техніки.
- Форма для додавання нових даних про витрату пального для вибраної техніки.
- Статистика та прогноз машинного навчання для вибраної техніки.

Забезпечте можливість інтерактивного взаємодії з графіками та можливість додавання нових даних безпосередньо через дашборд.
